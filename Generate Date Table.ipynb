```python
from datetime import datetime
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.functions import rand
```

```python
def dates(start, end):
    df = pd.DataFrame({"Date": pd.date_range(start, end)})
    df["Day"] = df.Date.dt.weekday
    df["MonthKey"] = df.Date.dt.month
    df["Month"] = df['MonthKey'].astype(str).str.zfill(2)
    df["Quarter"] = df.Date.dt.quarter
    df["Year"] = df.Date.dt.year
    df["YearMonth"] = df["Year"].astype(str) + '-' + df["Month"].astype(str)
    
    return df
```

```python
datePd = dates('2024-01-01', str(datetime.today().year + 2) + '-01-01')
```

```python
datesDf = spark.createDataFrame(datePd)
datesDf.write.mode("overwrite").format("delta").saveAsTable("Dates")
```
